```python
import pandas as pd

inventory_sets = pd.read_csv("inventory_sets.csv")
inventories = pd.read_csv("inventories.csv")
inventory_parts = pd.read_csv("inventory_parts.csv")
```

```python
inventory_sets

```

```python
#standardize column names
def clncolnm(df):
    df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
    return df

inventory_sets = clncolnm(inventory_sets)
inventories = clncolnm(inventories)
inventory_parts = clncolnm(inventory_parts)

#check for duplicates
print("Duplicates:")
print("Inventory Sets:", inventory_sets.duplicated().sum())
print("Inventories:", inventories.duplicated().sum())
print("Inventory Parts:", inventory_parts.duplicated().sum())

#drop duplicates
inventory_sets.drop_duplicates(inplace=True)
inventories.drop_duplicates(inplace=True)
inventory_parts.drop_duplicates(inplace=True)
```

```python
#merge inventory_parts with inventories
merged = pd.merge(inventory_parts, inventories, left_on="inventory_id", right_on="id", how="left")

#merge the result with inventory_sets
merged = pd.merge(merged, inventory_sets, on="inventory_id", how="left", suffixes=('_part', '_set'))


```
